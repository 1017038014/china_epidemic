```python
import requests
from pyecharts import Map, Line
city_url = 'http://ncovdata.spbeen.com/apis/get_china_citys/?province='
province_url = 'http://ncovdata.spbeen.com/apis/get_china_provinces/'
province_resp = requests.get(province_url)
data = province_resp.json()
data = data['data']
province_name = []
for d in data:
    province_name.extend(d.keys())
print(province_name,len(province_name)-1)



for province in province_name:
    temp_url = city_url + province
    temp_resp = requests.get(temp_url)
    cdata = temp_resp.json()['data']
    citys_name = []
    citys_confirmed = []
    citys_suspected = []
    citys_cured = []
    citys_dead = []
    for cd in cdata:
        k = "".join(cd.keys())
        citys_name.append(k)
        citys_confirmed.append(cd[k].get('confirmed',0))
        citys_suspected.append(cd[k].get('suspected',0))
        citys_cured.append(cd[k].get('cured',0))
        citys_dead.append(cd[k].get('dead',0))
        
    print(citys_name)
    print(citys_confirmed)
    print(citys_suspected)
    print(citys_cured)
    print(citys_dead)
    print('--------------'*10)
    
    if len(citys_name)<1 and len(citys_confirmed)<1 and len(citys_suspected)<1 and len(citys_cured)<1 and len(citys_dead)<1:
        print("跳过{}地区\n".format(province))
        continue
    
    pmap = Map(province+"地图", province+'疫情地图', width=1000, height=650)
    
    pmap.add("确诊", citys_name, citys_confirmed, maptype=province, visual_range=[0,max(citys_confirmed)], is_visualmap=True, visual_text_coloe='#000', is_label_show=True)
    
    pmap.add("疑似", citys_name, citys_suspected, maptype=province, visual_range=[0,max(citys_confirmed)], is_visualmap=True, visual_text_coloe='#000', is_label_show=True)
    
    pmap.add("治愈", citys_name, citys_cured, maptype=province,     visual_range=[0,max(citys_confirmed)], is_visualmap=True, visual_text_coloe='#000', is_label_show=True)
    
    pmap.add("死亡", citys_name, citys_dead, maptype=province,      visual_range=[0,max(citys_confirmed)], is_visualmap=True, visual_text_coloe='#000', is_label_show=True)
    
    pmap.render(path="省份地图/"+province+'-map.html')
```
