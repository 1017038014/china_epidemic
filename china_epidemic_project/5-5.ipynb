```python
import requests
url = 'http://ncovdata.spbeen.com/apis/get_china_provinces/?query_date='
date = '2020-02-'
target_data = { }
for i in range(1,15):
    tmp_url = url+date+str(i)
    resp = requests.get(tmp_url)
    result = resp.json()
    data = result['data']
    for d in data:
        k = "".join(d.keys())
        confirmed_num = d[k].get('confirmed',0)
        suspected_num = d[k].get('suspected',0)
        cured_num = d[k].get('cured',0)
        dead_num = d[k].get('dead',0)
        
        
        tdict = target_data.get(k,{})

        
        confirmed = tdict.get('confirmed',[])
        confirmed.append(confirmed_num)
        tdict['confirmed'] = confirmed
        
        suspected = tdict.get('suspected',[])
        suspected.append(suspected_num)
        tdict['suspected'] = suspected
        
        cured = tdict.get('cured',[])
        cured.append(cured_num)
        tdict['cured'] = cured
        
        dead = tdict.get('dead',[])
        dead.append(dead_num)
        tdict['dead'] = dead

        tdate = tdict.get('date',[])
        tdate.append(date+str(i))
        tdict['date'] = tdate

        target_data[k] = tdict

for key in target_data.keys():
    print(key, target_data[key],'\n')
    
    
    
from pyecharts import Map, Bar, Line, Page
for key in target_data.keys():
    print(key, target_data[key],'\n')
    x_line = target_data[key]['date']
    qz = target_data[key]['confirmed']
    ys = target_data[key]['suspected']
    zy = target_data[key]['cured']
    sw = target_data[key]['dead']
    line = Line(key+'疫情走势图')
    line.add('确诊', x_line, qz, is_label_show=True)
    line.add('疑似', x_line, ys, is_label_show=True)
    line.add('治愈', x_line, zy, is_label_show=True)
    line.add('死亡', x_line, sw, is_label_show=True)
    line.render(path="省份疫情走势图/"+key+"疫情走势图（2020年2月份1号至14号）.html")
```
