```python
from pyecharts import Map
import random
```

```python
from province_city import province_city 
```

```python
province_name = province_city.keys()
```

```python
province_value = [ ]
for i in range(len(province_name)):
    province_value.append(random.randint(1,1000))
```

```python
china = Map("中国地图",'中国地图',width=800, height=600)
```

```python
china.add("测试画出中国地图", province_name, province_value, maptype='china', visual_range=[0,1000], is_visualmap=True, visual_text_color='#000', is_label_show=True)
```
