```python
import requests
url = 'http://ncovdata.spbeen.com/apis/get_china_provinces/?query_date=2020-02-13'
resp = requests.get(url)
data = resp.json()
data = data['data']

province_name = []
confirmed_all = []
suspected_all = []
cured_all = []
dead_all = []
for d in data:
    k = "".join(d.keys())
    cscd = d[k]
    #print(cscd)
    if cscd != {}:
       confirmed = cscd['confirmed']
       suspected = cscd['suspected']
       cured = cscd['cured']
       dead = cscd['dead']
       #print(k, confirmed, suspected, cured, dead)
       province_name.append(k)
       confirmed_all.append(confirmed)
       suspected_all.append(suspected)
       cured_all.append(cured)
       dead_all.append(dead)
    else:
       pass
print(province_name)
print(confirmed_all)
print(suspected_all)
print(cured_all)
print(dead_all)
```

```python
from pyecharts import Map
china = Map("中国地图",'中国疫情地图【2020-02-13】',width=800, height=600)
china.add("各省确诊人数", province_name, confirmed_all, maptype='china', visual_range=[0,max(confirmed_all)//20], is_visualmap=True, visual_text_color='#000', is_label_show=True)
china.add("各省治愈人数", province_name, cured_all, maptype='china', visual_range=[0,max(confirmed_all)//20], is_visualmap=True, visual_text_color='#000', is_label_show=True)
china.add("各省死亡人数", province_name, dead_all, maptype='china', visual_range=[0,max(confirmed_all)//20], is_visualmap=True, visual_text_color='#000', is_label_show=True)
```

```python
china.render(path='2020-02-13中国各省疫情互动地图展示.html')
```
